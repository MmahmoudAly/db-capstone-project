```python
#Task 3
# the full name and contact details for every customer that has placed an order greater than $60
query = """SELECT CONCAT(c.GuestFirstName, ' ', c.GuestLastName) AS full_name, c.CustomerPhone, c.mail 
FROM CustomerDetails c
JOIN Orders o ON c.CustomerID = o.CustomerID
WHERE o.Cost > 60"""
cursor.execute(query)
results = cursor.fetchall()
for row in results:    
    print(f"Full Name: {row[0]}, Phone: {row[1]}, Email: {row[2]}")
```
